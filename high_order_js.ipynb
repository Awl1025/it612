```python
%%writefile code.js

const data = [5, 12, 3, 20, 11, 9, 30];

const result = [];
for (let i = 0; i < data.length; i++) {
  if (data[i] > 10) {
    result.push(data[i] * 2);
  }
}

console.log(result);



const numbers = [1, 2, 3, 4, 5, 6];

const result2 = numbers
  .filter(n => n % 2 === 0)   // keep evens
  .map(n => n * 2)           // double them
  .reduce((sum, n) => sum + n, 0); // sum

console.log(result2);



const users = [
  { id: 1, name: "Alice", age: 25 },
  { id: 2, name: "Bob", age: 17 },
  { id: 3, name: "Carol", age: 19 }
];


const hasUnder18 = users.some(user => user.age < 18);
console.log(hasUnder18); // true

const userWithId3 = users.find(user => user.id === 3);
console.log(userWithId3); // { id: 3, name: "Carol", age: 19 }

users.forEach(user => {
  console.log(user.name);
});



function repeat(n, action) {
  for (let i = 0; i < n; i++) {
    action(i);
  }
}

// Tests
repeat(3, console.log); // 0, 1, 2
repeat(4, i => console.log("Step " + i));
```
